```python
import pandas as pd
import re
```

```python
df = pd.read_csv("/kaggle/input/data-science-job-posting-on-glassdoor/Uncleaned_DS_jobs.csv")
```

```python
df.drop('index', inplace=True, axis=1)
df.head()
```

```python
df['Salary Estimate'].unique()
```

```python
df['Salary Estimate'] = df['Salary Estimate'].str.split('(').str[0]
df['Salary Estimate'] = df['Salary Estimate'].str.replace('K', '').str.replace('$', '')
df[['min_salary', 'max_salary']] = df['Salary Estimate'].str.split('-', expand=True).astype(int)
```

```python
df["min_salary"].unique()
```

```python
df["max_salary"].unique()
```

```python
df['avg_salary'] = (df['min_salary'] + df['max_salary']) // 2
```

```python
df['Company Name'] = df['Company Name'].str.replace('\n.*', '')
```

```python
import numpy as np
df.Rating = np.where(df.Rating==-1.0,0,df.Rating)
```

```python
df.loc[df.Rating == -1.0, 'Rating'] = 0
```

```python
df['job_state'] = df['Location'].str.split(',').str[-1].str.strip()
```

```python
df['same_state'] = (df['Location'] == df['Headquarters']).astype(int)
```

```python
current_year = 2023
df['company_age'] = current_year - df['Founded'].where(df['Founded'] >= 1, 0)
```

```python
df['Job Description'][2].split('\n')
```

```python
keywords = ['python', 'excel', 'hadoop', 'spark', 'aws', 'tableau', 'big data']

# Create binary columns for each keyword
for keyword in keywords:
    df[keyword] = df['Job Description'].str.lower().apply(lambda x: 1 if keyword in x else 0)
```

```python
df.head()
```

```python
def title_simplifier(title):
    title = title.lower()
    keyword_mapping = {
        'data scientist': 'data scientist',
        'data engineer': 'data engineer',
        'analyst': 'analyst',
        'machine learning': 'mle',
        'manager': 'manager',
        'director': 'director'
    }
    return keyword_mapping.get(title, 'na')
```

```python
def seniority(title):
    title = title.lower()
    senior_keywords = ['sr', 'senior', 'lead', 'principal']
    if any(keyword in title for keyword in senior_keywords):
        return 'senior'
    elif 'jr' in title:
        return 'jr'
    else:
        return 'na'
```

```python
df["Seniority"] = df["Job Title"].apply(seniority)
```

```python
df["Job Title"] = df["Job Title"].apply(title_simplifier)
```

```python
df = df[~((df['job_state']=='Remote') | (df['job_state']=='Utah') | (df['job_state']=='New Jersey') | (df['job_state']=='Texas') | (df['job_state']=='California'))]
df.job_state = df.job_state.apply(lambda x: x.replace('United States','US'))
df.job_state.value_counts()
```
